```python
import numpy as np 
def deserialize_shap_payload(shap_string):
    """
    Désérialise les données SHAP stockées au format string avec séparateurs.
    """
    parts = shap_string.split(';')
    if len(parts) != 4:
        raise ValueError("Format de données SHAP invalide")
    
    shap_values = np.array([float(v) for v in parts[0].split('|')])
    base_value = float(parts[1])
    feature_names = parts[2].split('|')
    feature_values = np.array([float(v) for v in parts[3].split('|')])
    
    return shap_values, base_value, feature_names, feature_values
```
