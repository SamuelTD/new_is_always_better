```python
import numpy as np 
def deserialize_shap_payload(shap_string):
    """
    Désérialise les données SHAP stockées au format string avec séparateurs.
    """
    parts = shap_string.split(';')
    if len(parts) != 4:
        raise ValueError("Format de données SHAP invalide")
    
    shap_values = np.array([float(v) for v in parts[0].split('|')])
    base_value = float(parts[1])
    feature_names = parts[2].split('|')
    feature_values = np.array([float(v) for v in parts[3].split('|')])
    
    return shap_values, base_value, feature_names, feature_values
```

```python
import pandas as pd
df = pd.DataFrame({'feature_names': feature_names, 
                   'feature_values': feature_values,
                   'shap_values': shap_values})

df['abs_shap'] = np.abs(df['shap_values'])
df_sorted = df.sort_values('abs_shap', ascending=False)
```

```python
import matplotlib.pyplot as plt
import shap
import seaborn as sns

# Sélection des 10 plus importantes
top_df = df_sorted.iloc[:10]

# Somme des autres
others = df_sorted.iloc[10:]
others_sum = others['shap_values'].sum()
others_row = pd.DataFrame({
    'feature_names': ['Autres'],
    'feature_values': [None],
    'shap_values': [others_sum],
    'abs_shap': [np.abs(others_sum)]
})

# Fusion des deux
final_df = pd.concat([top_df, others_row], ignore_index=True)

# Graphique
plt.figure(figsize=(12, 8))
plt.title('Impact des features sur la prédiction (Top 10 + Autres)')
colors = ['green' if x > 0 else 'red' for x in final_df['shap_values']]
plt.barh(range(len(final_df)), final_df['shap_values'], color=colors)
plt.yticks(range(len(final_df)), final_df['feature_names'])
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('Impact SHAP (Contribution à la prédiction)')
plt.gca().invert_yaxis()  # Inverser l'ordre des y pour que le plus grand soit en haut
plt.tight_layout()
plt.show()
```

```python
top_n = 15  # Montrer les 15 plus importantes features
top_features = df_sorted.head(top_n)

plt.figure(figsize=(10, 6))
sns.barplot(x='abs_shap', y='feature_names', data=top_features, palette='viridis')
plt.title(f'Top {top_n} features les plus influentes')
plt.xlabel('Importance absolue (|SHAP value|)')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```
